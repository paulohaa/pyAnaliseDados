# O que é um Jupyter Notebook?

Neste caso, o notebook é um formato de documento que contém tanto códigos quanto elementos de texto, figuras, links, vídeos, equações, etc. Como podemos misturar texto e código em um mesmo lugar, esses documentos são lugares ideais para gerar e descrever análises onde tanto a formulação da ideia, a visualização do resultado e a descrição do resultado pode ser feita num mesmo lugar.

# Células

Há dois tipos principais de células: 

- **[Markdown](https://help.github.com/articles/basic-writing-and-formatting-syntax/)**: células, como esta, utilizadas para descrever o que está sendo implementado. Elas utilizam alguns caracteres especiais para formatar o texto, como `** texto em negrito **` para fazer **texto em negrito**;
- **Code**: trechos de código para executar os comandos necessários para a análise. No nosso caso, os comandos serão escritos em Python 3, mas o Jupyter Notebook tem suporte para outras linguagens também. Mostraremos alguns comandos simples aqui, mas se quiser saber um pouco mais, segue [uma referência às regras básicas da linguagem](http://www.devfuria.com.br/python/sintaxe-basica/).

## Células de código (code)

Cada célula é um ambiente isolado que pode ser rodado separadamente. Você pode rodar cada célula individualmente e na sequência que desejar, mas as variáveis que ela usa precisam estar referenciadas antes do seu uso.

```python
# Esta é uma célula de código 
for i in range(0, 5):
    print(i)
```

## Além de processar, o código pode ser usado para visualizar dados

O código a seguir utiliza o pacote `matplotlib` para exibir um gráfico simples. 

Na área de imports colocamos um `%matplotlib inline`, certo? Essa é uma das "mágicas" disponíveis no Jupyter, que permite que os gráficos do `matplotlib` sejam apresentados dentro do próprio notebook, como podemos ver abaixo.

```python
# Esta é uma célula de código

import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(range(0,100))
```

E, agora, combinando com `numpy` para ter uma curva mais elaborada.

Além disso, são adicionados alguns metadados ao gráfico que, além de ser exibido, é salvo no arquivo `teste.png`.

```python
import numpy as np

# Montando a curva, com duas sequências de números - a segunda é derivada da primeira por uma fórmula
x = np.arange(0.0, 2.0, 0.01)
y = np.sin(2*np.pi*x)
plt.plot(x, y)

# Inserindo metadados e formatação
plt.xlabel('X (raio)')
plt.ylabel('Y (seno da circunferência de raio X)')
plt.title('Uma curva um pouco mais sofisticada')
plt.grid(True)

# Salvando em arquivo
plt.savefig("teste.png", format="png")
```
